# Case Study Activity: Smart Traffic Management System

## Problem Statement

A smart city uses intelligent traffic devices. Every device receives the same command activate(), but each device performs a different task.

### Requirements

1. Create a parent class called TrafficDevice.

2. Create the following child classes:
   - TrafficLight
   - SpeedCamera
   - PedestrianSignal

3. Each child class must override the activate() method.

4. Create one object from each class, store them in a list, and activate each device without checking its type.

5. Add another child class called EmergencySiren without modifying the activation loop.

### Objective

Demonstrate the concept of Polymorphism in Object-Oriented Programming by allowing different objects to respond differently to the same method call.

```
class TrafficDevice:
    def activate(self):
        print("Traffic device activated.")


class TrafficLight(TrafficDevice):
    def activate(self):
        print("Traffic Light: Green light is now ON.")


class SpeedCamera(TrafficDevice):
    def activate(self):
        print("Speed Camera: Monitoring vehicle speed.")


class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("Pedestrian Signal: Walk sign displayed.")


class EmergencySiren(TrafficDevice):
    def activate(self):
        print("Emergency Siren: Emergency vehicle approaching.")


devices = [
    TrafficLight(),
    SpeedCamera(),
    PedestrianSignal(),
    EmergencySiren()
]

print("Activating Smart Traffic Devices...\n")

for device in devices:
    device.activate()
```
